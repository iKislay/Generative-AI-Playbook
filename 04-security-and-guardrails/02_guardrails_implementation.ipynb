Learn how to implement defensive layers using modern guardrail frameworks.

### Topics Covered:

2. **Output Validation** using Pydantic and Guardrails AI
3. **PII Redaction** before sending to LLM
4. **Semantic Routing** for sensitive queries

```python
from pydantic import BaseModel, Field, validator

class AgentOutput(BaseModel):
    thought: str = Field(description="Internal reasoning of the agent")
    action: str = Field(description="The tool to call")
    action_input: dict = Field(description="Parameters for the tool")

    @validator('action')
    def validate_action(cls, v):
        allowed_actions = ['web_search', 'read_file']
        if v not in allowed_actions:
            raise ValueError(f"Action {v} is unauthorized!")
        return v

print("Structural Guardrail defined. Agent cannot execute unauthorized tools.")
```
