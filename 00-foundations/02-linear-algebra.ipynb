# Linear Algebra for AI

## 1. Overview
Linear Algebra is the language of modern AI. From representing images as pixels to encoding words as high-dimensional vectors, every piece of data is treated as a mathematical entity. This lesson covers the building blocks of data representation: scalars, vectors, matrices, and tensors.

## 2. Why this exists
Computers don't understand text or images directly; they only understand numbers. Linear Algebra provides the tools to organize these numbers and perform complex transformations on them efficiently, which is the basis for how neural networks learn.

## 3. Problem statement
How do we represent a paragraph of text or a 4K video in a way that a machine can process? How do we calculate the similarity between two sentences or update billions of model parameters simultaneously?

## 4. Intuition
- **Vector:** Think of it as a list of features (e.g., [Height, Weight, Age]).
- **Matrix:** A collection of vectors, like a spreadsheet or a digital image.
- **Tensor:** A multi-dimensional grid. A color image is a 3D tensor (Width x Height x RGB Channels).
- **Dot Product:** A way to measure how much two vectors "align" or how similar they are.

## 5. Visual explanation
### Data Dimensions

| Rank | Name | Example | Shape |
| :--- | :--- | :--- | :--- |
| 0 | Scalar | A single number (Temperature) | () |
| 1 | Vector | A list of numbers (Coordinates) | (n,) |
| 2 | Matrix | A table (Grayscale Image) | (m, n) |
| 3 | Tensor | A cube (Color Image) | (m, n, 3) |
| 4 | Tensor | A batch of color images | (b, m, n, 3) |

## 6. Math
**Matrix Multiplication:** If matrix A is (m x n) and matrix B is (n x p), their product C is (m x p). This operation is the core of every "Layer" in a transformer model.

**Dot Product Formula:** 
Sum of products of corresponding elements: `A · B = ∑ (a_i * b_i)`

## 7. Architecture
In AI, we use **Tensors** as the universal data structure. Modern hardware like GPUs and TPUs are specifically designed to perform matrix multiplications much faster than standard CPUs.

## 8. Step-by-step flow
1. **Vectorization:** Convert raw data (text/images) into numbers.
2. **Linear Transformation:** Multiply the input vector by a weight matrix.
3. **Activation:** Apply a non-linear function to the result.
4. **Aggregation:** Sum up the results to produce a prediction.

## 9. Code implementation


```python
import numpy as np

# Create two vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Dot product (Similarity measure)
dot_product = np.dot(v1, v2)
print(f"Dot Product: {dot_product}")

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = np.matmul(A, B)
print(f"Matrix Product:\n{C}")
```

## 10. Production considerations
- **Precision:** High-precision math (FP64) is slow. Most AI models use FP16 or BF16 to save memory and speed up computation.
- **Sparsity:** Many matrices in AI have lots of zeros. Sparse matrix operations can save massive amounts of compute power.
- **Quantization:** Reducing 32-bit numbers to 8-bit or 4-bit for faster inference on edge devices.

## 11. Common mistakes
- **Dimension Mismatch:** Trying to multiply a (3x2) matrix by a (3x2) matrix. The inner dimensions must match (e.g., 3x2 and 2x4).
- **Transposition Errors:** Forgetting to transpose a weight matrix before multiplication.
- **Broadcasting Confusion:** Assuming NumPy will always handle different shapes the way you expect.
