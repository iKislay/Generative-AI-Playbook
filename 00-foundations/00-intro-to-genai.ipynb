## 4. Intuition
AI is not a single invention but a progression of layers:
- **Artificial Intelligence (AI):** The broad field of building machines that mimic human intelligence.
- **Machine Learning (ML):** Instead of hard-coding rules, machines learn patterns from data (e.g., recommendation systems).
- **Deep Learning (DL):** A subset of ML that uses neural networks as universal approximators.
- **Transformers:** Introduced in 2017, they changed how models handle sequences like text, enabling long-range dependencies and scalability.
- **Generative AI (GenAI):** Models that create content. Transformers unlocked today's breakthroughs in language generation.

**What makes LLMs "Large"?**
1. **Large datasets:** Web-scale corpora spanning billions of words.
2. **Large numbers of parameters:** Billions or trillions of tunable weights.
3. **Large computing requirements:** Massive computing power needed for training.

At their core, LLMs are next-word predictors. Given "I'd like to eat a loaf of ____", the model assigns a higher probability to "bread" than "noodles" based on learned contextual dependencies.

## 7. Architecture
The Training Pipeline of Large Language Models involves four key stages:
1. Unsupervised Pre-Training
2. Supervised Fine-Tuning
3. Reinforcement Learning with Human Feedback (RLHF)
4. Chat Optimisation

## 8. Step-by-step flow
- **Stage 1 (Pre-Training):** The model learns to predict the next word from unstructured data, becoming a "foundation model".
- **Stage 2 (Fine-Tuning):** The model is trained on curated instruction-response data (prompt -> response) to follow human directions.
- **Stage 3 (RLHF):** Human annotators rank responses. A reward model is trained to guide the LLM to align with human intent and safety guidelines.
- **Stage 4 (Chat Optimisation):** The model is fine-tuned for multi-turn conversations, role conditioning, and context memory.

## 11. Common mistakes
**Confusing Generative AI with Agentic AI:**
- **Generative AI** focuses on creating content based on patterns.
- **Agentic AI** reasons, plans, and acts toward a goal with minimal human intervention. 
Agentic AI does not inherently depend on Generative AI, but modern agentic systems often use GenAI as an enhancement for communication, reasoning, and content generation.
