# Statistics for AI

## 1. Overview
Statistics provides the tools to understand data. While probability helps us predict future events based on known rules, statistics helps us discover the rules by analyzing past data. In AI, we use statistics to evaluate datasets, understand model performance, and ensure our results are significant.

## 2. Why this exists
When training an AI, you are working with millions of data points. You cannot manually inspect every image or text snippet. Statistics allows you to summarize massive datasets, identify outliers, and measure how well your model generalizes to new data.

## 3. Problem statement
How do we know if our dataset is biased? How can we prove that Model A is truly better than Model B, and not just getting lucky on a specific test set?

## 4. Intuition
Imagine you want to know the average height of people in a city. You can't measure everyone. Instead, you measure a representative sample (statistics) and infer the true average. AI works the same way: it learns patterns from a sample dataset to make predictions about the real world.

## 5. Visual explanation
### Key Statistical Concepts

| Concept | Definition | AI Application |
| :--- | :--- | :--- |
| **Mean / Median** | Central tendency of data | Finding the average loss during training |
| **Variance / Std Dev** | How spread out the data is | Normalizing input features |
| **Correlation** | Relationship between variables | Feature selection (finding what matters) |
| **P-Value** | Measure of statistical significance | A/B testing different model prompts |

## 6. Math
**Standardization (Z-score):** `z = (x - mean) / standard_deviation`
This formula is used to scale features so they have a mean of 0 and a standard deviation of 1, which helps neural networks learn faster.

## 7. Architecture
Statistical principles are embedded in the training loop. Functions like **Loss** (e.g., Mean Squared Error or Cross Entropy) are statistical measures of how far the model's predictions are from the truth.

## 8. Step-by-step flow
1. **Exploratory Data Analysis (EDA):** Calculate mean, median, and variance of your dataset.
2. **Normalization:** Scale your data using standard deviation.
3. **Training:** The model minimizes the statistical error (loss) over the dataset.
4. **Evaluation:** Calculate metrics like Precision, Recall, and F1-Score to judge performance.

## 9. Code implementation


```python
import numpy as np

# Sample dataset: house prices in thousands
prices = np.array([250, 300, 320, 280, 500, 310, 290])

# Basic statistics
mean_price = np.mean(prices)
median_price = np.median(prices)
std_dev = np.std(prices)

print(f"Mean: {mean_price:.2f}")
print(f"Median: {median_price:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# Notice how the outlier (500) affects the mean more than the median
```

## 10. Production considerations
- **Data Drift:** Over time, the statistical properties of real-world data change (e.g., user behavior shifts). Models must be retrained to adapt.
- **Bias:** If your training data is statistically biased (e.g., unrepresentative demographics), your model will amplify that bias.
- **Evaluation Metrics:** Accuracy is not always the best metric. Use Precision, Recall, or F1-Score for imbalanced datasets.

## 11. Common mistakes
- **Ignoring Outliers:** Training a model on data with extreme outliers can skew the results drastically.
- **P-Hacking:** Manipulating data or tests just to achieve statistical significance.
- **Confusing Mean and Median:** Using the mean when the data is highly skewed (like income).
