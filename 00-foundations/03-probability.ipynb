# Probability for AI

## 1. Overview
Probability is the foundation of decision-making in AI. Large Language Models (LLMs) do not "know" facts in the human sense; they calculate the probability of the next word in a sequence. This lesson covers the core concepts that allow machines to handle uncertainty.

## 2. Why this exists
The real world is messy and full of uncertainty. Probability provides a mathematical framework to reason about things we aren't 100% sure about. In AI, it helps us choose the most likely answer given a set of inputs.

## 3. Problem statement
Given the sentence "The cat sat on the...", how does an AI decide between "mat," "floor," or "refrigerator"? How does it measure its own confidence in an answer?

## 4. Intuition
Think of probability as a "degree of belief." If you flip a coin, you have a 50% belief it will be heads. In AI, the model has a "belief" across all possible words in its vocabulary, and it usually picks the one with the highest belief (probability).

## 6. Math
**Bayes Theorem:** `P(A|B) = [P(B|A) * P(A)] / P(B)`
This tells us how to update the probability of a hypothesis (A) based on new evidence (B). 

**Softmax Function:** Converts a list of raw scores into a probability distribution where all numbers sum up to 1.

## 7. Architecture
The final layer of almost every classification AI or LLM is a **Softmax Layer**. It takes the output of the neural network and turns it into a list of probabilities for each possible output category.

## 8. Step-by-step flow
1. **Input:** The model receives a sequence of tokens.
2. **Scoring:** The model calculates a "logit" score for every word in its vocabulary.
3. **Normalization:** The Softmax function turns these scores into probabilities.
4. **Sampling:** The model chooses the next word based on these probabilities (sometimes using techniques like Top-K or Top-P sampling).

## 9. Code implementation


```python
import torch
import torch.nn.functional as F

# Raw scores (logits) for 3 possible words
logits = torch.tensor([2.0, 1.0, 0.1])

# Convert to probabilities using Softmax
probabilities = F.softmax(logits, dim=0)

words = ["mat", "floor", "refrigerator"]
for word, prob in zip(words, probabilities):
    print(f"Word: {word:12} Probability: {prob.item():.4f}")

print(f"Sum of probabilities: {probabilities.sum().item()}")
```

## 10. Production considerations
- **Temperature:** A parameter that controls how "creative" or "focused" a model is. High temperature (e.g., 1.2) makes low-probability words more likely, while low temperature (e.g., 0.1) makes the model very predictable.
- **Calibration:** Ensuring that if a model says it is 90% sure, it is actually correct 90% of the time.
- **Hallucinations:** When a model assigns high probability to a sequence of words that is factually incorrect.

## 11. Common mistakes
- **Ignoring the Prior:** Forgetting that some events are naturally very rare, even if current evidence seems to support them.
- **Overconfidence:** Deep learning models are often overconfident in their predictions, especially on data they haven't seen before.
- **Confusing Correlation with Causation:** Just because A and B often happen together doesn't mean A causes B.
